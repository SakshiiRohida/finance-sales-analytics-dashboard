```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/finance_sales_project/cleaned_sales_data.csv")
```

```python
df["discount_band"].value_counts()
```

```python
df.groupby("discount_band")["sales"].mean().plot(kind="bar")
plt.title("Average Sales by Discount Band")
plt.show()
```

```python
df.groupby("discount_band")["profit"].mean().plot(kind="bar")
plt.title("Average Profit by Discount Band")
plt.show()
```

```python
df["margin"] = df["profit"] / df["sales"]
df.groupby("discount_band")["margin"].mean().plot(kind="bar")
plt.title("Average Margin by Discount Band")
plt.show()
```
